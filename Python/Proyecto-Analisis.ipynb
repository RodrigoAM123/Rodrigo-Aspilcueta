```python
import pandas as pd 

videojuegos = pd.read_csv("videogamessales.csv", sep=';')
videojuegos
```

```python
videojuegos.info()
#Como se ve la columna "yEAR" tiene como tipo de dato: decimal o float
```

```python
videojuegos["Year"]=videojuegos["Year"].astype(int)
videojuegos.head(5)
#Con esto cambiamos de FLOAT a int.
```

```python
videojuegos = videojuegos.rename(columns={
    'Name': 'Nombre',
    'Platform': 'Plataforma',
    'Year': 'AÃ±o',
    'Genre': 'GÃ©nero',
    'Publisher': 'Distribuidor',
    'NA_Sales': 'Ventas_NA',
    'EU_Sales': 'Ventas_EU',
    'JP_Sales': 'Ventas_JP',
    'Other_Sales': 'Ventas_Otros'
})
videojuegos.head(1)
#AcÃ¡ solo traduje los nombres de cada columna
```

```python
videojuegos['Ventas_Globales'] = videojuegos['Ventas_NA'] + videojuegos['Ventas_EU'] + videojuegos['Ventas_JP'] + videojuegos['Ventas_Otros']
videojuegos.head(5)
```

```python
videojuegos = videojuegos.sort_values(by='Ventas_Globales', ascending=False)
videojuegos['Ranking'] = range(1, len(videojuegos) + 1) 
videojuegos
#Ordene y agregÃ© el ranking para conocer que videojuego ganÃ³ mÃ¡s.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

top_10 = videojuegos.head(10)  # Selecciona el top 10

plt.figure(figsize=(12, 6))
sns.barplot(x='Ventas_Globales', y='Nombre', data=top_10, palette='viridis')

plt.xlabel('Ventas Globales (millones)')
plt.ylabel('Juego')
plt.title('Top 10 Juegos MÃ¡s Vendidos')
plt.show()
```

```python
ventas_por_genero = videojuegos.groupby('GÃ©nero')['Ventas_Globales'].sum().sort_values(ascending=False)

plt.figure(figsize=(8, 8))
ventas_por_genero.head(5).plot(kind='pie', autopct='%1.1f%%', startangle=90, colormap='Set2')

plt.title('Ventas globales por gÃ©nero (Top 5)')
plt.ylabel('')
plt.show()
```

```python
ventas_por_plataforma = videojuegos.groupby('Plataforma')['Ventas_Globales'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
ventas_por_plataforma.plot(kind='bar', color='skyblue')

plt.xlabel('Plataforma')
plt.ylabel('Ventas globales (millones)')
plt.title('Ventas globales por plataforma')
plt.xticks(rotation=45)
plt.show()
#AcÃ¡ se da a entender que la PS2 es la reina en ventas globales.
```

```python
ventas_por_aÃ±o = videojuegos.groupby('AÃ±o')['Ventas_Globales'].sum()

plt.figure(figsize=(12, 6))
plt.plot(ventas_por_aÃ±o.index, ventas_por_aÃ±o.values, marker='o', linestyle='-', color='red')

plt.xlabel('AÃ±o')
plt.ylabel('Ventas globales (millones)')
plt.title('EvoluciÃ³n de ventas de videojuegos')
plt.grid(True)
plt.show()

#AcÃ¡ un punto aparte porque estas ventas son con productos fÃ­sicos y no digitales. Y hay que recalcar que el mercado digital supero por
#mucho el mercado fÃ­sico.
```
